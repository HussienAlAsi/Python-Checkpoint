```python
#Question1
array = [1,2,3,5,8]
l = 0
h = len(array) - 1
def binary_search(array,l,h,x):
    if h >= l: 
        mid = int(l+(h-l)/2)
        if array[mid] == x:
            return "True"
        elif array[mid] > x:
            return binary_search(array,l,mid-1,x)
        elif array[mid] < x:
            return binary_search(array,mid+1,h,x)
    else:
        return "False"
print(binary_search(array,l,h,5))
print(binary_search(array,l,h,6))
```

```python
#Question2
#All bases to the power of 0 = 1, all bases to the power of 1 are equal to themselves and 0 to the power of anything is 0
# a^10 = a * a^9 == a^b = a * a*b-1
def power(a,b):
    if b == 0:
        return 1
    elif b == 1:
        return a
    elif a == 0:
        return 0
    else:
        return a*power(a,b-1)
print(power(3,4))
    
```

```python
#Question3
def bubblesort(list1):
    for n in range(len(list1)-1,0,-1):
        for i in range(n):
            if list1[i] > list1[i+1]:
                temp = list1[i]
                list1[i] = list1[i+1]
                list1[i+1] = temp 
        return list1 
list1 = [29,13,22,37,52,49,46,71,56]
print(bubblesort(list1))
```

```python
#Question4
def mergesort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left = arr[:mid]
        right = arr[mid:]
        
        
        mergesort(left)
        mergesort(right)
        
        i=j=k=0
        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                arr[k] = left[i]
                i += 1
            else:
                arr[k] = right[j]
                j += 1
            k += 1
                
                
        while i < len(left):
                arr[k] = left[i]
                i += 1
                k += 1
        while j < len(right):
                arr[k]=right[j]
                j += 1
                k += 1
            
arr = [29,13,22,37,52,49,46,71,56]
mergesort(arr)
print(arr)
```

```python
#Question5
def quicksort(array):
    if len(array) <2:
        return array
    else:
        pivot = array[0]
        less_arr = [i for i in array[1:] if i <= pivot]
        greater_arr = [i for i in array[1:] if i >= pivot]
        return quicksort(less_arr) + [pivot] + quicksort(greater_arr)
print(quicksort([29,13,22,37,52,49,46,71,56]))
```
