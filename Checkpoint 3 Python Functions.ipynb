```python
#Question 3
def Sum(a):
    b= sum(a)
    return(b)
def Multiply_items(c):
    x = 1
    for i in c:
        x = x * i
    return(x)
list1 = [1,2,3,4,5,7,8,9,10]
def odd_list(list1):
    odd_numbers = []

    for number in list1:
        if number % 2 == 1 :
            odd_numbers.append(number)

    return odd_numbers
def even_list(list1):
    even_numbers = []
    for number in list1:
        if number % 2 == 0:
            even_numbers.append(number)
            return even_numbers
list_odd = odd_list(list1)
list_even = even_list(list1)
result = Sum(list_even) + Multiply_items(list_odd)
print(result)



    


    




```
