```python
import time
t = int(input("Please Enter Time in Seconds : "))
def countdown(t):
    while t: 
        mins, secs = divmod(t, 60)
        timeformat = '{:02d}:{:02d}'.format(mins, secs)
        print(timeformat, end = '\r')
        time.sleep(1)
        t -=1
    print('Fire in the hole')
    
    
countdown(t)
        
        
```
